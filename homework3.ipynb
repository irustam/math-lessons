## Задание 3
Напишите код на Python, реализующий построение графиков:
- окружности,
- эллипса,
- гиперболы.


```python
import numpy as np
from matplotlib import pyplot as plt
from pylab import rcParams
```

Окружность

```python
R = 100
x = []
y = []
y2 = []
for i in range(-1000, 1000):
    x.append(i)
    y.append(np.sqrt(R**2 - i**2))
    y2.append(-np.sqrt(R**2 - i**2))

rcParams["figure.figsize"] = 5, 5
plt.plot(x, y, color="red")
plt.plot(x, y2, color="red")
plt.xlabel('x')
plt.ylabel('y')
```

Элипс

```python
a = 100
b = 50

x = []
y = []
y2 = []
for i in range(-1000, 1000):
    x.append(i)
    y.append(np.sqrt(1 - i**2 / a**2) * b)
    y2.append(-np.sqrt(1 - i**2 / a**2) * b)

rcParams["figure.figsize"] = 6, 6
plt.axis([-110, 110, -110, 110])
plt.plot(x, y, color="red")
plt.plot(x, y2, color="red")
plt.xlabel('x')
plt.ylabel('y')
```

Гипербола

```python
a = 9
b = 7

x = []
y = []
y2 = []
for i in range(-1000, 1000):
    x.append(i)
    y.append(np.sqrt(i**2 / a**2 - 1) * b)
    y2.append(-np.sqrt(i**2 / a**2 - 1) * b)

rcParams["figure.figsize"] = 6, 6
plt.axis([-50, 50, -50, 50])
plt.plot(x, y, color="red")
plt.plot(x, y2, color="red")
plt.xlabel('x')
plt.ylabel('y')
```

1. Нарисуйте трехмерный график двух параллельных плоскостей.


```python
from mpl_toolkits.mplot3d import Axes3D
from pylab import figure
from pylab import show
```

```python
fig = figure()
ax = Axes3D(fig)
A = -5
B = -6
C = 4
D = 2

X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
Z = (-A*X - B*Y - D)/C

D2 = 40
Z2 = (-A*X - B*Y - D2)/C

ax.plot_wireframe(X, Y, Z, color='yellow')
ax.plot_wireframe(X, Y, Z2, color='red')

show()
```

## Задание 4
1. Решите систему уравнений:
#### y = x2 – 1
#### exp(x) + x∙(1 – y) = 1

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-5, 5, 201)
rcParams["figure.figsize"] = 8, 5
plt.plot(x, x**2 - 1)
plt.plot(x, 1 - (1 - np.exp(x))/ x)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

```python
def equation(p):
    x, y = p
    return (y - x**2 + 1, 1 - (1 - np.exp(x))/ x - y)

x1, y1 = fsolve(equation, (1, 1))
print(x1, y1)
```
